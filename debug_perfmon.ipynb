```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pprint import pprint
```

```python
counter_file = "C:\\PerfLogs\\Admin\HCImage stream to disk\AHRENS-LS2_20200810-000004\\D-drive_logicaldisk_memory_processor.csv"
counter_filename = os.path.basename(counter_file).split('.')[0]
drive = "D:"
savepath = drive + "\\jing\\HCImageLive\\"
```

```python
df = pd.read_csv(counter_file)
df_init = df.iloc[0]
df = df.iloc[1:]
df.head()
```

```python
counters = list(df.keys())
for counter in counters:
    print(counter)
```

```python
drive_counters = [counter for counter in counters if drive in counter]
for counter in drive_counters:
    print(counter)
```

```python
write_counters = [counter for counter in counters if (drive in counter) and ("Write" in counter)]
for counter in write_counters:
    print(counter)
```

```python
df.loc[:,f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Disk Write MB/sec'] = df.loc[:, f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Disk Write Bytes/sec'].astype('float') /1024/1024
```

```python
counters = list(df.keys())
write_counters = [counter for counter in counters if (drive in counter) and ("Write" in counter)]
for counter in write_counters:
    print(counter)
```

```python
write_df = df.filter(write_counters)
write_df.head()
```

```python
axes = write_df.astype(np.float).plot(subplots=True,
                                      figsize=(9.5, 2*len(write_counters)),
)
plt.savefig(savepath + counter_filename + '.png')
```

```python
df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\% Free Space'].plot()

## Not the "Free Megabytes" I imagined!
# df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Cumulative Megabytes'] = df_init[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Free Megabytes'] - df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Free Megabytes']
# df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Written Megabytes'] = np.diff(np.insert(df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Cumulative Megabytes'].to_numpy(),0,0))
```

```python
df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Written Megabytes'] = df.loc[:, f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Disk Write MB/sec'].cumsum()
df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Written Megabytes'].plot()
```

```python
halfTB = 1000000 * 0.75
written_halfTB = np.where(df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Written Megabytes'] > halfTB)[0][0]  # 1TB/2
starting = np.where(df[f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Written Megabytes'] <= 0)[0][-1]
duration = written_halfTB - starting
```

```python
axes = df.filter([f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Written Megabytes',f'\\\\AHRENS-LS2\\LogicalDisk({drive})\\Disk Write MB/sec']).plot(subplots=True,
 title=['cumulative written', 'write speed'],
 legend=False)

for ax in axes:
    ax.axvline(x=starting, color='k', linestyle='--')
    ax.axvline(x=written_halfTB, color='k', linestyle='--')

axes[0].axhline(y=halfTB,color='k', linestyle='--')
axes[0].set_ylabel('MB')
axes[1].set_ylabel('MB/s')
axes[1].set_xlabel('Time [s]')

axes[1].set_ylim(top=1200)
axes[1].set_xlim(right=2000)
plt.suptitle(f'{drive} drive time to 0.75 TB = {duration}s')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.savefig(savepath + counter_filename + 'summary.png')
```
